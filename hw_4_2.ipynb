```python
import hashlib
# Данные пользователей (логины и хэши паролей)
users_data = {
    "user1": hashlib.sha256(b"password123").hexdigest(),
    "user2": hashlib.sha256(b"mypassword").hexdigest(),
}
def hash_password(password):
    """
    Функция для хэширования пароля
    """
    return hashlib.sha256(password.encode('utf-8')).hexdigest()
def login_user(username, password):
    """
    Функция для авторизации пользователя
    """
    password_hash = hash_password(password)
    if username in users_data:
        if users_data[username] == password_hash:
            return "Авторизация успешна"
        else:
            return "Неверный пароль"
    else:
        return "Пользователь не найден"
# Ввод данных пользователем
user_username = input("Введите логин: ")
user_password = input("Введите пароль: ")
# Попытка авторизовать пользователя
login_result = login_user(user_username, user_password)
print(login_result)
```
